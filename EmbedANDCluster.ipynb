```python
parameters=['full','tied','diag','spherical']
n_clusters=np.arange(1,10)
results_=pd.DataFrame(columns=['Covariance Type','Number of Cluster','Silhouette Score','Davies Bouldin Score'])
for i in parameters:
    for j in n_clusters:
        gmm_cluster=GaussianMixture(n_components=j,covariance_type=i,random_state=123)
        clusters=gmm_cluster.fit_predict(e)
        if len(np.unique(clusters))>=2:
          results_=results_.append({"Covariance Type":i,'Number of Cluster':j,"Silhouette Score":metrics.silhouette_score(e,clusters),
                                    'Davies Bouldin Score':metrics.davies_bouldin_score(e,clusters)}
                                   ,ignore_index=True)
```
