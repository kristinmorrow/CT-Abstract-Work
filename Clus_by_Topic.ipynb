```python
hd = df[df['summarized_sent'].str.contains('|'.join(HD))].set_index("abstract_id")
#del hd['index']
del hd['Unnamed: 0']
#hd = hd[hd['summarized_sent'].str.contains('|'.join('exercise'))]
hd
```

```python
data1 = []
k = 100
for i in (range(0, len(hd), k)):
    embed = hd.iloc[i:i+k]
    print(i, embed.iloc[0].name, embed.iloc[-1].name)
    data1.append(model1.encode(embed['summarized_sent'].tolist()))
```

```python
h = np.concatenate(data1, axis=0 )
h.shape
```

Title: Segmentation of Credit Card Users in Python

Author: MuhammetBektas

Availability: https://github.com/muhammetbektas/Unsupervised-Learning/blob/master/Segmentation_of_Credit_Card_Users.ipynb

```python
n_clusters=np.arange(2,5)
results_=pd.DataFrame(columns=['Covariance Type','Number of Cluster','Silhouette Score','Davies Bouldin Score'])
for j in n_clusters:
  gmm_cluster=GaussianMixture(n_components=j,covariance_type='full',random_state=123)
  clusters=gmm_cluster.fit_predict(h)
  if len(np.unique(clusters))>=2:
    results_=results_.append({'Number of Cluster':j,"Silhouette Score":metrics.silhouette_score(h,clusters),
                                    'Davies Bouldin Score':metrics.davies_bouldin_score(h,clusters)}
                                   ,ignore_index=True)
```

```python
display(results_.sort_values(by=["Silhouette Score"], ascending=False)[:10])
```

```python
# 2 clusters ss:.151
#display(results_.sort_values(by=["Silhouette Score"], ascending=False)[:10])
gmm_cluster=GaussianMixture(n_components=4,covariance_type="full",random_state=13)
gmm_cluster.fit(h)
gmm_labels = gmm_cluster.predict(h)
hd['hd_clust'] = gmm_labels
hd
```

```python
hd_clus_0 = hd[hd['hd_clust']==0] 
hd_clus_1 = hd[hd['hd_clust']==1] 
hd_clus_2 = hd[hd['hd_clust']==2] 
hd_clus_3 = hd[hd['hd_clust']==3] 
hd_clus_0 = hd_clus_0['summarized_sent'].tolist()
hd_clus_1 = hd_clus_1['summarized_sent'].tolist()
hd_clus_2 = hd_clus_2['summarized_sent'].tolist()
hd_clus_3 = hd_clus_3['summarized_sent'].tolist()
```
